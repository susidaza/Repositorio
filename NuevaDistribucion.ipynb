Generación de muestras de una variable aleatoria normal usando numpy.

```python
import numpy as np

x2 = 7 #Grados de libertad
n=1000 #Numero de replicas
vals = np.random.chisquare(x2,n)
print(vals)
```

Cálculo de estadísticas usando pandas.

```python
import pandas as pd

df = pd.DataFrame(vals)
print(df.describe())
```

Generación de histograma usando matplotlib.

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(x=vals, bins=30)
plt.title('Histograma de tiempos de servicio')
plt.xlabel('Tiempos de servicio')
plt.ylabel('Frecuencia')
plt.show()

print(bins)
print(count)
```

Densidad de Chi cuadrado con 7 grados de libertad

```python
from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt

x= np.arange(0,30,0.001)
plt.plot(x, chi2.pdf(x,x2))
plt.title('Función de densidad')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

Densidad Chi cuadrado con 3 grados de libertad

```python
from scipy.stats import chi2
import numpy as np
import matplotlib.pyplot as plt

x= np.arange(0,30,0.001)
plt.plot(x, chi2.pdf(x,3))
plt.title('Función de densidad normal(mu=3, sigma=0.5)')
plt.xlabel('Valores')
plt.ylabel('Densidad')
plt.show()
```

QQ plot usando statsmodels. 

```python
import statsmodels.api as sm
import scipy.stats as stats
fig = sm.qqplot(data = vals, dist=stats.chi2, fit = False, distargs=(x2,), line="45")
plt.show()
```
